```python
import xarray as xr
from matplotlib import pyplot as plt
from matplotlib import colors
%matplotlib inline 
```

```python
f = xr.open_dataarray("TS_cmems_reanalysis_temperature_from1987to2019_AdriaticSea.nc")
f
```

```python
f.plot.line(color="r",linewidth=3, figsize=(18,5))
plt.xlabel('Time Series')
plt.ylabel('Sea Surface Temperature (C)')
plt.title('Sea Surface Temperature from 1987 to 2019 over Adriatic Sea');
```

```python
fy_dt = f.groupby('time.year').mean()
fy_dt
```

```python
fy_dt.plot.line(color="k",linewidth=3,marker='o', figsize=(18,5))
plt.xlabel('Time Series by year')
plt.ylabel('Sea Surface Temperature (C)')
plt.title('Mean Sea Surface Temperature from 1987 to 2019 over Adriatic Sea');
```

```python
##### mean of 12 months for all 33 years

fm_dt = f.groupby('time.month').mean()
fm_dt
```

```python
#plt.plot(fm_dt,color='k',linewidth=3,marker='o')
fm_dt.plot.line(color="k",linewidth=3,marker='o', figsize=(18,5))
plt.xlabel('Time Series by month')
plt.ylabel('Sea Surface Temperature (C)')
plt.title('Mean Sea Surface Temperature from 1987 to 2019 over Adriatic Sea');
```

```python
roller_m = fm_dt.rolling(month=12,center=True)
roller_m
```

```python
roller_m.mean()
```

```python
roller_mm=roller_m.construct("window_dim")
#print(roller_mm)
```

```python
plt.plot(roller_mm,color='k',linewidth=3,marker='o')
plt.plot(fm_dt, color='r', label='monthly rolling 12',linewidth=3)
plt.legend();
plt.xlabel('Time Series by month')
plt.ylabel('Sea Surface Temperature (C)')
plt.title('Rolling Mean=12 Sea Surface Temperature by month from 1987 to 2019 over Adriatic Sea');
```

```python
roller_mm.plot(cmap=plt.cm.jet)

plt.xlabel('Time Series by month')
plt.ylabel('Sea Surface Temperature (C)')
plt.title('Rolling Mean=12 Sea Surface Temperature by month from 1987 to 2019 over Adriatic Sea');
```

```python
roller_y = fy_dt.rolling(year=33,center=True)
roller_y
```

```python
roller_yy=roller_y.construct("window_dim")
```

```python
plt.plot(roller_yy,color='k',linewidth=3,marker='o')
plt.plot(fy_dt, color='r', label='year rolling 33',linewidth=3)
plt.legend();

plt.xlabel('Time Series by year')
plt.ylabel('Sea Surface Temperature (C)')
plt.title('Rolling Mean=33 Sea Surface Temperature by year from 1987 to 2019 over Adriatic Sea');
```
