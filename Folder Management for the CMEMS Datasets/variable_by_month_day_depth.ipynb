```python
days, datasets = zip(*nut_ds[j].groupby('time.day'))
paths = ['./2014_MONTHLY_thetao_BYDEPTH/2014-01-%sT12:00:00_temperature_AdriaticSea_Z10_CF.nc' % y for y in days]
print(paths)
xr.save_mfdataset(datasets, paths)
```
