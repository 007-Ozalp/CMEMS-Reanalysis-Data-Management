```python
import plotly.express as px
import xarray as xr
import os
import sys
import numpy as np
import xarray as xr
import matplotlib.pyplot as plt
import pandas as pd
import plotly.graph_objects as go
import pandas as pd
import dash
import dash_core_components as dcc
import dash_html_components as html
from dash import Dash, dcc, html, Input, Output  

import pandas as pd

df = pd.read_csv('2D_1day1.csv')


```

```python
fig = go.Figure(data=go.Scattergeo(
    lat = df['lat'][:],
    lon = df['lon'][:],
    text = df['time'] + '' + df['thetao'].astype(str) + ' T[C]',

    marker = dict(
        color = df['thetao'],
        reversescale = False,
        opacity = 1,
        size = 2,
        colorbar = dict(
            titleside = "right",
            outlinecolor = "rgba(68, 68, 68, 0)",
            ticks = "outside",
            #showticksuffix = "last",
            dtick = 1
        )
    )
))


fig=fig.update_layout(
    geo = dict(
        scope = 'europe',
        resolution = 50,
        showcoastlines=True, coastlinecolor="RebeccaPurple",
        showland=True, landcolor="LightGreen",
        showocean=True, oceancolor="LightBlue",
        showlakes=True, lakecolor="Blue",
        showrivers=True, rivercolor="Blue",
        projection = dict(
            type = 'natural earth',
            
            rotation_lon = None
        ),
        lonaxis = dict(
            showgrid = True,
            gridwidth = 0.5,
            range= [ 12, 22 ],
            dtick = 3
        ),
        lataxis = dict (
            showgrid = True,
            gridwidth = 0.5,
            range= [ 37, 46 ],
            dtick = 3
        )
    ),
    
   



)
fig.update_layout(height=300, margin={"r":0,"t":0,"l":0,"b":0})

app = dash.Dash(__name__)
app.layout = html.Div(
    [html.H1("16 November 2020 @5 m: \n Temperature Sea PHY Analysis"), dcc.Graph(figure=fig),]
)
if __name__ == "__main__":
    app.run_server()
```
